```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
data=pd.read_csv("Student_Marks.csv")
data.head()
display(data.head())
```

```python
data.columns
```

```python
x = data[['time_study', 'number_courses']]   # input features
y = data['Marks']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print("Predicted Marks:", y_pred)
print("Actual Marks:", y_test.values)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error:", mae)
print("R² Score:", r2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot of time_study vs Marks
plt.figure(figsize=(6,4))
sns.scatterplot(x=x_train['time_study'], y=y_train)
plt.title("Study Time vs Marks")
plt.xlabel("Study Time (hours)")
plt.ylabel("Marks")
plt.show()

# Scatter plot of number_courses vs Marks
plt.figure(figsize=(6,4))
sns.scatterplot(x=x_train['number_courses'], y=y_train)
plt.title("Number of Courses vs Marks")
plt.xlabel("Number of Courses")
plt.ylabel("Marks")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Create model
model = LinearRegression()

# Train model
model.fit(x_train, y_train)

# Print coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

```python
# Predict on test data
y_pred = model.predict(x_test)

# Compare actual vs predicted
df_compare = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
display(df_compare.head())

# Evaluate performance
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R² Score:", r2_score(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, y_pred, color='blue')
plt.xlabel("Actual Marks")
plt.ylabel("Predicted Marks")
plt.title("Actual vs Predicted Marks")
plt.show()

```
